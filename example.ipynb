```python
import os

def split_file(file_path, chunk_size=1*1024*1024):
    with open(file_path, 'rb') as f:
        while True:
            chunk = f.read(chunk_size)
            if not chunk:
                break
            yield chunk


path = "audio/2_5201837958239356469.mp3"
# Example usage
for i, chunk in enumerate(split_file(path)):
    with open(f'audio/output_{i}.mp3', 'wb') as f:
        f.write(chunk)
```

```python
from pydub import AudioSegment
from pydub.silence import split_on_silence
#reading from audio mp3 file

path = "audio/2_5201837958239356469.mp3"

sound = AudioSegment.from_mp3(path)
# spliting audio files
audio_chunks = split_on_silence(sound, min_silence_len=500, silence_thresh=-40 )
#loop is used to iterate over the output list
for i, chunk in enumerate(audio_chunks):
   output_file = "audio/2_5201837958239356469/chunk{0}.mp3".format(i)
   print("Exporting file", output_file)
   chunk.export(output_file, format="mp3")
```

```python
from pydub import AudioSegment

path = "audio/2_5201837958239356469.mp3"

song = AudioSegment.from_mp3(path)

# PyDub handles time in milliseconds
ten_minutes = 10 * 60 * 1000

first_10_minutes = song[:ten_minutes]

first_10_minutes.export("good_morning_10.mp3", format="mp3")
```
